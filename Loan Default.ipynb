# 2.Data Cleaning #

```python
#Since the employment duration units were not give, i took it to be in terms of Hours worked and in this lines of code i am converting it to years by dividing it by 24 and then by 365
df["Employment Duration"] = (df["Employment Duration"] / 24)/365                    
df["Employment Duration"]
df["Employment Duration"] = df["Employment Duration"].astype(int)
df["Employment Duration"]
```

```python
#Dropping all columns accept: Loan Status, Interest Rate, Debit to Income, Employment Duration, Loan Amount, Delinquency - two years, Grade, Revolving Utilities, Total Current Balance and Verification Status
# Loan Status is the target variable, Interest rate shows risk pricing, Debit to Income shows financial pressure, Employment Duration shows stability, Loan Amount shows exposure level, Delinquency - two years shows historical payment behavior, Grade shows bank risk classification, Revolving Utilities shows credit stress, Total Current Balance shows existing liabilities and Verification Status shows information reliability
columns = ["Loan Status","Interest Rate","Debit to Income","Employment Duration","Loan Amount","Delinquency - two years","Grade","Revolving Utilities","Total Current Balance","Verification Status","Home Ownership"]
for column in df.columns:
    if column not in columns:
        df.drop(column, axis=1, inplace=True)

df.head()
```

# 4. Diagnostic Analysis

```python
# Question 1: What is Grade's influence on loan default?
grade_risk = (
    pd.crosstab(
        df["Grade"],
        df["Loan Status"],
        normalize="index"
    )*100
)

grade_risk.sort_values(
    by=1,
    ascending=False
)
```

### Finding

Lower loan grades exhibited higher default rates, and the chi-square test confirms the relationship between loan grade and loan outcome is statistically significant. This shows that borrower risk classification is a strong predictor of default.

```python
from scipy.stats import chi2_contingency
table = pd.crosstab(
    df["Grade"],
    df["Loan Status"]
)

chi2_contingency(table)
```

```python
#Question 2: Does Interest Rate have an influence on loan default?
df.groupby("Loan Status")["Interest Rate"].mean()
```

### Finding

The average interest rate was higher for defaulted loans, and the t-test indicates this difference is statistically significant. This suggests that borrowers charged higher interest rates were more likely to default, consistent with risk-based pricing.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(
    x="Loan Status",
    y="Interest Rate",
    data=df
)

plt.title("Interest Rate by Loan Status")
plt.show()
```

```python
from scipy.stats import ttest_ind
#Here we perform a t-test to see if there is a significant difference in interest rates between defaulters and non-defaulters
default = df[df["Loan Status"] == 1]["Interest Rate"]
non_default = df[df["Loan Status"] == 0]["Interest Rate"]

ttest_ind(default, non_default)
```
